```python
import pandas as pd
star_wars = pd.read_csv("star_wars.csv", encoding="ISO-8859-1")
star_wars.head(10)
```

```python
star_wars = star_wars[pd.notnull(star_wars["RespondentID"])]
```

```python
yes_no = {
    "Yes": True,
    "No": False
}
seen = "Have you seen any of the 6 films in the Star Wars franchise?"
fan = "Do you consider yourself to be a fan of the Star Wars film franchise?"
star_wars[seen] = star_wars[seen].map(yes_no)
star_wars[fan] = star_wars[fan].map(yes_no)
```

```python
import numpy as np
mapping = {
    "Star Wars: Episode I  The Phantom Menace": True,
    np.nan: False,
    "Star Wars: Episode II  Attack of the Clones": True,
    "Star Wars: Episode III  Revenge of the Sith": True,
    "Star Wars: Episode IV  A New Hope": True,
    "Star Wars: Episode V The Empire Strikes Back": True,
    "Star Wars: Episode VI Return of the Jedi": True
}
i = 1
for column in star_wars.columns[3:9]:
    star_wars[column] = star_wars[column].map(mapping)
    star_wars = star_wars.rename(columns = {column: "seen_" + str(i)})
    i += 1
    
star_wars.head()    
```

```python
i = 1
for column in star_wars.columns[9:15]:
    star_wars[column] = star_wars[column].astype(float)
    star_wars = star_wars.rename(columns = {column: "ranking_" + str(i)})
    i += 1
```

```python
means = star_wars.mean()
%matplotlib inline
means[8:15].plot.bar(color = "blue")
```

### Findings:
So far we changed the column names, removed null data, and changed some of data for better readability.

We can say from the bar plot above, that episode 5 in the Star Wars series, i.e., "The Empire Strikes Back" is the most favorite movie, while episode 3 i.e., "Revenge of the Sith" is the least favorite movie. 

```python
sums = star_wars[star_wars.columns[3:9]].sum()
sums.plot.bar(color = "blue")
```

### Findings:
It's obvious that the most favorite Star Wars movie "The Empire Strikes Back" has been seen most by fans, while the least favorite movie has been seen least.

```python
males = star_wars[star_wars["Gender"] == "Male"]
females = star_wars[star_wars["Gender"] == "Female"]

import matplotlib.pyplot as plt

fig = plt.figure(figsize=(6, 6))
for sp in range(0, 4):
    ax = fig.add_subplot(2, 2, sp + 1)
    if sp==0 or sp==2:
        df = males
    else:
        df = females
    if sp==0 or sp==1:
        comp = df[df.columns[3:9]].sum()
    else:
        comp = df[df.columns[9:15]].mean()
    comp.plot.bar(color = "blue")
plt.show()
```

### Findings:
Although, there's no significant difference in rankings of the movies, the view counts have been lower for women than men. It might be because Star Wars is seen as male-oriented action-scifi genre for movies.

```python
import numpy
rank_edu = star_wars.groupby("Education").agg(numpy.mean)
rank_edu = rank_edu[rank_edu.columns[8:15]]
seen_edu = star_wars.groupby("Education").agg(numpy.sum)
seen_edu = seen_edu[seen_edu.columns[2:8]]
rank_edu.plot.bar(legend = None)
```

### Findings:
We can see that "Empire Strikes Back" movie has been the highest ranked movie in Star Wars series regardless of education and "Revenge of the Sith" has been the lowest ranked movie.

```python
seen_edu.plot.bar()
```

### Findings:
We can observe here that Star Wars movies have been mainly watched by people with some sort of college experience. Lower level of education puts significant economic strain on people to buy tickets for the movie. Also, "Empire Strikes Back" has been the most watched movie as expected.

```python
rank_loc = star_wars.groupby("Location (Census Region)").agg(numpy.mean)
rank_loc = rank_loc[rank_loc.columns[8:15]]
seen_loc = star_wars.groupby("Location (Census Region)").agg(numpy.sum)
seen_loc = seen_loc[seen_loc.columns[2:8]]
rank_loc.plot.bar(legend = None)
```

### Findings:
"Empire Strikes Back" strikes one more time as the most favorite movie regardless of location.

```python
seen_loc.plot.bar(legend = None)
```

### Findings:
East South Central region has the least number of seen movies, however, that may be due to fewer people from that region responded to survey.
